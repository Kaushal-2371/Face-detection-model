```python
import cv2
facedect=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
image=cv2.imread("Raw_img.jpg")
```

# Formating images

```python
from matplotlib import pyplot as plt
plt.imshow(image)
```

### By default it comes in this formate, you can convert it into RGB

```python
rgbimg=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(rgbimg)
```

### Before editing, make it Gray format

```python
grimage=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(grimage)
```

# Dectacting faces

```python
faces = facedect.detectMultiScale(
    grimage,                # Passing the Gray image
    scaleFactor=1.8,        # Giving the scale to search faces ((larger scale -> rough search) (smaller scale -> fine search))
    minNeighbors=2,         # Minimum faces to search, don't worry if you have given more
    minSize=(30,30)         # SIze of rectraingle
)
print(faces)
```

# Marking rectrangles

```python
image=cv2.imread("Raw_img.jpg")
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 5)
    #cv2.rectangle(image,(Starting point),(ending point),(color),width)
rgbimg=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(rgbimg)
```
